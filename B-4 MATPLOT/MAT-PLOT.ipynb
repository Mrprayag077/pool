```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df=pd.read_csv(r"C:\Users\ADMIN$\Desktop\pppppppp\PRACT-4 S FIRE\forestfires.csv", sep=',')
```

For more information, read [Cortez and Morais, 2007].

X - x-axis spatial coordinate within the Montesinho park map: 1 to 9
Y - y-axis spatial coordinate within the Montesinho park map: 2 to 9
month - month of the year: 'jan' to 'dec'
day - day of the week: 'mon' to 'sun'
FFMC - FFMC index from the FWI system: 18.7 to 96.20
DMC - DMC index from the FWI system: 1.1 to 291.3
DC - DC index from the FWI system: 7.9 to 860.6
ISI - ISI index from the FWI system: 0.0 to 56.10
temp - temperature in Celsius degrees: 2.2 to 33.30
RH - relative humidity in %: 15.0 to 100
wind - wind speed in km/h: 0.40 to 9.40
rain - outside rain in mm/m2 : 0.0 to 6.4
area - the burned area of the forest (in ha): 0.00 to 1090.84
(this output variable is very skewed towards 0.0, thus it may make
sense to model with the logarithm transform).

```python
# Define the columns to visualize
COLS = ['X', 'Y', 'month', 'day', 'FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH',
       'wind', 'rain']

# Set up the figure and axes wxh
fig, axes = plt.subplots(nrows=3, ncols=4, figsize=(12, 10))

# Plot the data
for col, ax in zip(COLS, axes.flatten()):
    sns.histplot(df[col], ax=ax)
    ax.set_title(col)

# Adjust the layout and spacing
plt.tight_layout()

# Display the plot
plt.show()
```
