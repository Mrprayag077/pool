```python
import requests      #standard for making HTTP requests in Python
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
link="https://www.flipkart.com/search?q=laptop&sid=6bo%2Cb5g&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_6_na_na_na&as-pos=1&as-type=RECENT&suggestionId=laptop%7CLaptops&requestId=7ec220e8-4f02-4150-9e0b-9e90cf692f4b&as-searchtext=laptop"
```

```python
page = requests.get(link)
```

```python
soup = bs(page.content, 'html.parser')
#it gives us the visual representation of data
print(soup.prettify())
```

```python
name=soup.find('div',class_="_4rR01T")
print(name)
```

```python
#get rating of a product
rating=soup.find('div',class_="_3LWZlK")
print(rating)
rating.text
```

```python
#get other details and specifications of the product
specification=soup.find('div',class_="fMghEO")
print(specification)
specification.text
```

```python
price=soup.find('div',class_='_30jeq3 _1_WHN1')
print(price)
price.text
```

```python
products=[]              #List to store the name of the product
prices=[]                #List to store price of the product
ratings=[]               #List to store rating of the product
```

```python
for data in soup.findAll('div',class_='_3pLy-c row'):
        name=data.find('div', attrs={'class':'_4rR01T'})
        price=data.find('div', attrs={'class':'_30jeq3 _1_WHN1'})
        rating=data.find('div', attrs={'class':'_3LWZlK'})
        
        
        products.append(name.text)

        prices.append(price.text)

        ratings.append(rating.text)
        
      
```

```python
df = pd.DataFrame({'Product Name':products,'Prices':prices,'Ratings':ratings})

df.shape
```
